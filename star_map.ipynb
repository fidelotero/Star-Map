### III. Create two columns ra and dec in the data frame having the right ascension in decimal degrees and declinations in decimal degrees.

````python
df['ra'] = (df['RAh'] + (df['RAm'] / 60) + (df['RAs'] / 3600)) * 15
````

### V. Create two columns long and lat that are the galactic longitude and galactic latitude. Using Astropy convert ra and dec into galactic longitude and galactic latitude.

````python
from astropy import units as u
from astropy.coordinates import SkyCoord
c = SkyCoord(ra=df['ra']*u.degree, dec=df['dec']*u.degree)
````

````python
df['long'] = c.galactic.l
````

````python
df['long'] = df['long']
````

````python
df['lat'] = c.galactic.b
````

````python
df['lat'] = df['lat']
````

### VI. Assign colors to the stars according to spectral types:

````python
colors_dict = {'O': '#7F00FF', 'B': '#0000cd', 'A': '#ADD8E6', 'F': '#008000',
              'G': '#FFFF00', 'K': '#FFA500', 'M': '#ff0000'}

df['colors'] = ''
for ind in df['SpType'].loc[df['SpType'].isna() == False].index:
    for char in df['SpType'][ind]:
        if char.isupper():
            df['colors'][ind] = char
            break
df['colors_hex'] = ''

for ind in df.index:
    if df['colors'][ind] in colors_dict.keys():
        df['colors_hex'][ind] = colors_dict[df['colors'][ind]]
    else:
        df['colors_hex'][ind] = '#000000'
````

### VIII. Create an all-sky map of the stars in the Yale Bright Star Catalog under Aitoff projection. Look at the sample code under Files -> HWK -> HWK4 -> sky_map.pdf. Make sure that your maps have titles and axis labels.

##### Make an all-sky plot in equatorial coordinates for epoch 2000.0.
##### Make an all-sky plot in galactic coordinates.
##### Notice the difference in the view of the sky in the two coordinate systems.

````python
# all-sky plot in equatorial coordinates for epoch 2000.0
ra = c.ra.wrap_at(180 * u.deg).radian
dec = c.dec.radian

plt.figure(figsize=(20, 20))
plt.subplot(projection='aitoff')
plt.title('Map of the Sky')
plt.xlabel('Right Ascension')
plt.ylabel('Declination')
plt.scatter(ra, dec, 1, alpha=0.4, c=df.colors_hex)
plt.grid(True)
plt.savefig('sky_map_equatorial.pdf')
plt.show()
````

````python
# all-sky plot in galactic coordinates for epoch 2000.0
plt.figure(figsize=(20, 20))
plt.subplot(projection='aitoff')
plt.title('Map of the Sky')
plt.xlabel('Galactic Longitude')
plt.ylabel('Galactic Latitude')
plt.scatter(c.galactic.l.wrap_at(180 * u.deg).radian, c.galactic.b.radian, 1, alpha=0.4, c=df.colors_hex)
plt.grid(True)
plt.savefig('sky_map_galactic.pdf')
plt.show()
````
